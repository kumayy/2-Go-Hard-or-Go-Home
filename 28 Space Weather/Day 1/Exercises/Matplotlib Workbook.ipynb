### Line plots

```python
x = np.linspace(0, 2, 100)
```

```python
plt.figure?
```

```python
# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig = plt.figure(figsize=(12,10))
ax = plt.subplot(111)  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

fig = figsize=(10,8)
# red dashes, blue squares and green triangles

ax = plt.subplot(121)
ax.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
ax.grid()

ay = plt.subplot(122)
ay.plot(t, t, 'r--', label='t')
ay.plot(t, t**2, 'bs', label='$t^{2}$')
ay.plot(t, t**3, 'g^', label='$t^{3}$')
ay.grid()
ay.legend(loc=1)
```

### Scatter Plots

```python
np.random.randint?
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
data
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111)
cplot = ax.scatter('a', 'b', c='c', s='d', data=data)
cbar = plt.colorbar(cplot, ax=ax, shrink=0.9)
cbar.ax.set_ylabel('colour of b')
plt.xlabel('entry a')
plt.ylabel('entry b')

plt.show()
```

### Contour Plots

```python
x = np.linspace(-10,10,21)
y = np.linspace(0,20,21)

X, Y = np.meshgrid(x,y)

Z = np.random.rand(21,21)*100.
```

```python
print(np.shape(X))
```

```python
c_levels = np.linspace(Z.min(), Z.max(), 10)
print(c_levels)
```

```python
c_levels = np.linspace(0, 100, 11)
print(c_levels)
```

```python
fig = figsize=(10,8)
ax = plt.subplot(111)
cplot = ax.contourf(X, Y, Z, levels=c_levels, cmap='plasma')
cbar = plt.colorbar(cplot, ax=ax)
cbar.ax.set_ylabel('Z')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

### Polar Plots

```python
rad = np.linspace(0,10,11)
theta = np.linspace(0, 2*np.pi, 91)

R, T = np.meshgrid(rad,theta)

Z = np.zeros(np.shape(R))
```

```python
Z
```

```python
Z[R<2] = 20.
Z[R>6] = 60.
```

```python
Z
```

```python
fig = figsize=(10,8)
ax = plt.subplot(111, polar=True)
cplot = ax.contourf(T, R, Z, levels=c_levels, cmap='inferno')
cbar = plt.colorbar(cplot, ax=ax)
cbar.ax.set_ylabel('Z')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
fig = figsize=(10,8)
ax = plt.subplot(111, polar=True)
cplot = ax.contourf(T, R, Z, levels=c_levels, cmap='inferno')
cbar = plt.colorbar(cplot, ax=ax, pad=0.1)
cbar.ax.set_ylabel('Z', fontsize=12, weight='bold')
cbar.ax.tick_params(labelsize=12)
ax.set_theta_zero_location('S')
ax.set_xticks(np.deg2rad([0,90,180,270]))
ax.set_xticklabels(['00','06','12','18'], fontsize=12, weight='bold')
ax.set_yticks([0,2,4,6,8,10])
ax.set_yticklabels(['o', '$88^{\circ}$', '$86^{\circ}$', '$84^{\circ}$', '$82^{\circ}$', '$80^{\circ}$'], fontsize=12, weight='bold',color='white')
ax.set_title('Polar Contour Plot', fontsize=12, weight='bold')
```
