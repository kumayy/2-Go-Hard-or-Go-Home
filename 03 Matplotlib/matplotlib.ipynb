```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(0,10)
y = 2*x
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.title("String Title")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.xlim(0,6)
plt.ylim(0,12)
# plt.savefig("firstplot.png")
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = x*2
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Large figure
axes2 = fig.add_axes([0.1, 0.4, 0.5, 0.5]) # Smaller figure

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(x, y)
axes2.set_title('Small Figure');
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Large figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Smaller figure

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots(1,2)
#           OR
```

```python
fig, axes = plt.subplots(nrows= 1, ncols= 2)
axes[0].plot(a, b)
axes[1].plot(x, y)
```

```python
fig, axes = plt.subplots(nrows= 2, ncols= 2)
axes[0][0].plot(a, b)
axes[1][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][1].plot(x, y)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows= 2, ncols= 2, figsize=(4,2.5), dpi=180)
axes[0][0].plot(a, b)
axes[1][0].plot(x, y)
axes[1][0].set_xlabel("X_label 1,0")
axes[0][1].plot(a, b)
axes[1][1].plot(x, y)
axes[1][1].set_title("TITLE")
fig.suptitle("SUPER TITLE")

fig.subplots_adjust(wspace=0.5, hspace= 0.6)

fig.savefig("new_plot.png", bbox_inches="tight")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label="X vs X")
ax.plot(x,x**2, label="X ve X^2")
ax.legend(loc= (1.1,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x,x**2, color="purple",ls="-.")
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# custom dash
lines = ax.plot(x, x+8, color="black", lw=5)
# format: line length, space length
lines[0].set_dashes([1, 1,1,1,10,10])
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# Use marker for string code
# Use markersize or ms for size
ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) 
#ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) 
# make linewidth zero to see only markers
ax.plot(x, x-4,marker='2',ms=20)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

# marker size and color
ax.plot(x, x, color="black", lw=1, ls='-',
 marker='s', markersize=20, 
 markerfacecolor="red", markeredgewidth=8,
 markeredgecolor="blue");
```
