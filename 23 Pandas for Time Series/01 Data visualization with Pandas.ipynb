# Pandas Data Visualization

```python
df1 = pd.read_csv('df1.csv', index_col=0)
df2 = pd.read_csv('df2.csv')
```

### Scatter Plot

```python
df1.plot.scatter(x="A", y="B")
```

```python
df1.plot.scatter(x="A",y="B",c="C", cmap="CMRmap")
```

```python
df1.plot.scatter(x="A",y="B", s=df1["C"]*30, alpha =0.4)
```

```python
df1['C'].min()
```

```python
def add_three(val):
    return val+3

df1.plot.scatter(x='A',y='B',s=df1['C'].apply(add_three)*25, alpha=0.4);
```

```python
df2.plot.scatter(x='a',y='b',s=df2['c']*200).autoscale(enable=True, axis='both', tight=True);
```

### Hexagonal Bin Plot

### Title and Axis Labels

```python
df2.plot.line(figsize=(8,3));
```

```python
title =  "Custom Pandas Table"
ylabel = "Y - Axis Data"
xlabel = "X - Axis Data"
```

```python
ax = df2.plot.line(figsize=(8,3), title=title)
ax.set(xlabel=xlabel, ylabel=ylabel)
ax.autoscale(axis="x", tight=True);
```
