```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Scatter Plot

```python
df = pd.read_csv("dm_office_sales.csv")
df
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.scatterplot(x="salary",y="sales",
 data=df, hue="level of education", size="work experience",
 palette="magma", alpha=0.6, style="level of education")
```

## Distribution Plots

```python
df.head()
```

```python
df.info()
```

```python
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.set(style="whitegrid")
sns.histplot(data=df, x="salary", bins=20,
color="blue", edgecolor="black", linewidth=3, kde=True)
```

```python
sns.kdeplot(data=df, x="salary")
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages=pd.DataFrame(sample_ages, columns=["age"])
```

```python
sns.displot(data=sample_ages, x="age", rug=True,
 bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x="age",
clip=[0,100], bw_adjust=0.8)
```

## Categorical Plots

```python
df["division"].value_counts()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x="division", hue="level of education")
# plt.ylim(90,260)
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.barplot(data=df, x="level of education",
y="salary", estimator=np.mean, ci="sd", hue="division")
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
df=pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',
 y='math score',data=df,hue='gender')
# Optional move the legend outside
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',
 y='math score',data=df,hue='gender',split=True)
```

```python
sns.swarmplot(x='math score',y='race/ethnicity',data=df,size=3)
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x='race/ethnicity',
 y='math score',data=df,hue='gender')
```

## Comparison Plot

```python
sns.jointplot(data=df, x="math score",y="reading score",
 kind="hex")
```

```python
sns.jointplot(data=df, x="math score",y="reading score",
 kind="hist")
```

```python
sns.jointplot(data=df, x="math score",y="reading score",
 kind="kde", shade=True)
```

```python
sns.jointplot(data=df, x="math score",y="reading score",
 hue="gender")
```

```python
sns.pairplot(data=df, hue="gender", corner=True)
```

## Grids

```python
sns.catplot(data=df, x="gender",y="math score",
 kind="box", col="lunch", row="test preparation course")
```

```python
grid = sns.PairGrid(df, hue="gender")
grid = grid.map_upper(sns.scatterplot)
grid = grid.map_lower(sns.kdeplot)
grid = grid.map_diag(sns.histplot)
grid = grid.add_legend()
```

## Matrix Plot

```python
df = pd.read_csv('country_table.csv')
df
```

```python
df = df.set_index("Countries")
df
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop("Life expectancy", axis=1),
 lw= 0.5, annot=True, cmap="viridis")
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop("Life expectancy", axis=1),
 lw= 0.5, annot=True, cmap="viridis", col_cluster=False)
```
